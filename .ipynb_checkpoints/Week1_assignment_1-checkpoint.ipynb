```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df_steps = pd.read_csv('steps.csv', sep=';')
# df_steps['step_average'] = df_steps.mean(axis=1)
date_frame=list(df_steps.columns.values)
# df_steps_columns_mean = df_steps["20-6-2013":"13-5-2014"].mean()
df_steps_columns_mean = df_steps.loc[:,"20-6-2013":"13-5-2014"].mean()
print(df_steps_columns_mean)

```

```python
df_survey = pd.read_csv('survey.csv')
df_survey['height'].values[df_survey['height'] > 300] = 0
df_survey['weight'].values[df_survey['weight'] > 300] = 0
df_survey['steps_average'] = df_steps.mean(axis=1)
df_survey['body_mass'] = df_survey['weight']/ (df_survey['height']**2/10000)
df_survey['body_mass'] = df_survey['body_mass'].fillna(0.0).astype(int)
df_survey['steps_average'] = df_survey['steps_average'].fillna(0.0).astype(int)
df_survey.head(5)

```

```python
df_survey['body_mass'].name = 'Body mass distributions' #Labeling the variable for plotting
file_size = df_survey['body_mass'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(file_size, kde=True) 
plt.title('Body mass')
plt.show()
```

```python
# calculate some statistics for body mass

male_bm = df_survey.loc[df_survey["gender"] == "Male", "body_mass"]
female_bm = df_survey.loc[df_survey["gender"] == "Female", "body_mass"]
print('median for Male : ' + str(float(male_bm.median())))
print('median for Female : ' + str(float(female_bm.median())))
print('Overall median: ' + str(float(df_survey[['body_mass']].median())))
print('mode: ' + str(df_survey[['body_mass']].mode()['body_mass'][0]))
print('mean: ' + str(float(df_survey[['body_mass']].mean())))
print('standard deviation: ' + str(float(df_survey[['body_mass']].std())))
print('variance: ' + str(float(df_survey[['body_mass']].var()))) 
```
