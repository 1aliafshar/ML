## Import libraries 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr 
```

## Getting the data

```python
# open the dataframes with the right separators
df_steps = pd.read_csv('steps.csv', sep=';')
df_survey = pd.read_csv("survey.csv")
df = df_survey.merge(df_steps, on = "id")
df.head()
```

## Meaning of step among date


```python
df["steps_mean"] = df.loc[:,"20-6-2013":"13-5-2014"].mean(axis=1)
df.head()
```

Select the height, weight and steps variable to find a relationship between them

```python
df_selected = df[["city", "gender", "age","education_1", "height", "weight", "steps_mean"]]
df_selected.head()
```

print scatterplot matrix of H,W and steps

```python
sns.pairplot(df_selected)
plt.show()
```

plot all variable individually to find best mean value in each 

```python
sns.distplot(df_selected["height"], hist=True, kde=True)
plt.title("Height distribution of participants")
plt.xlabel("Height")
plt.ylabel("Proportion")
plt.show()

sns.distplot(df_selected["weight"], hist=True, kde=True)
plt.title("weight distribution of participants")
plt.xlabel("weight")
plt.ylabel("Proportion")
plt.show()


sns.distplot(df_selected["steps_mean"], hist=True, kde=True) 
plt.title("Steps distribution of participants")
plt.xlabel("Steps")
plt.ylabel("Proportion")
plt.show()
```

## Minify and remove extra data which too far from mean 

```python
df_selected = df_selected.drop(df_selected[(df_selected.weight > 200)].index)
df_selected = df_selected.drop(df_selected[(df_selected.steps_mean > 25000)].index)

sns.pairplot(df_selected)
plt.show()

sns.distplot(df_selected["steps_mean"], hist=True, kde=True) 
plt.title("Steps distribution of participants")
plt.xlabel("Steps")
plt.ylabel("Proportion")
plt.show()
```

## Plot Relation between height and wight with steps 


```python
sns.lmplot(x='height', y='weight', fit_reg=False, data=df_selected)
plt.xlabel('Height in cm') 
plt.ylabel('Weight in kg')
plt.title('Weight by height')

plt.show()
```

```python
sns.lmplot(x='weight', y='steps_mean', fit_reg=False, data=df_selected)
plt.xlabel('weight in cm') 
plt.ylabel('Mean steps per day')
plt.title('Mean steps by weight')
plt.show()
```

```python
sns.lmplot(x='height', y='steps_mean', fit_reg=False, data=df_selected)
plt.xlabel('Height in cm') 
plt.ylabel('Mean steps per day')
plt.title('Mean steps by height')
plt.show()
```

The correlation between height, weight and steps

```python
df_selected = df_selected.dropna() #I will drop the rows with NA, otherwise pearsonr won't work
print(pearsonr(df_selected['height'], df_selected['steps_mean']))
print(pearsonr(df_selected['height'], df_selected['weight']))
```

## Plot relation between a categorical and numerical variable


```python
df_selected = df_selected.drop(df_selected[(df_selected.steps_mean > 25000)].index)

df_selected = df_selected[["gender", "steps_mean"]].dropna() 
sns.catplot(data = df_selected, x = "gender", y = "steps_mean", kind="swarm") #swarm plot
plt.xlabel("Gender") 
plt.ylabel("Mean steps per day")
plt.title("Mean steps by gender")

sns.catplot(data = df_selected, x = "gender", y = "steps_mean", kind = "box") #box plot
plt.xlabel("Gender") 
plt.ylabel("Mean steps per day")
plt.title("Mean steps by gender")
plt.show()
```

## Select two different categorical variable to match

```python
pd.crosstab(df_selected["age"], df_selected["gender"], margins=True)
```

Normalize the age and gender 

```python
pd.crosstab(df_selected["age"], df_selected["gender"],normalize="columns", margins=True)
```

Visualize distirbution age bu gender on Barchart
